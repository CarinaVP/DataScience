la distribución tiene colas más pesadas que una distribución normal, lo que indica una distribución leptocúrtica (más puntiaguda en el centro)

Matriz de correlación: correlación de precio de venta con las demás

```python
corr_mat= df_train.corr()
```

```python
corr_mat
```

```python
import seaborn as sns ##graficas mas bonitas
```

```python
f, ax= plt.subplots(figsize=(12,9))
sns.heatmap(corr_mat, vmax=.8, square=True)       ##Que sea al cuadrado             
```

```python
k=10
k
```

```python
cols= corr_mat.nlargest(k,'SalePrice')['SalePrice'].index ## 10 variables mas correlacionadas con 'SalePrice'
```

```python
cols
```

```python
cm= np.corrcoef(df_train[cols].values.T)
```

```python
cm
```

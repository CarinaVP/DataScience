```python
#Utilizando la funcion apply
def millones(x):
  return x/1000000

paises3["pob en millones"]=paises3.poblacion.apply(millones)
paises3
```

```python
#agregando una columna al Dataframe "paises" que contenga la poblacion por cada km2 de superficie
def densidad(df):
  return df["poblacion"]/df ["sup. en km2"]


paises3.apply(densidad, axis=1) #con 1 lo bbusca el valor de la columna "poblacion" y sup en km2
```

```python
#Agregandolo al Dataframe original
def densidad(df):
  return df["poblacion"]/df ["sup. en km2"]


paises3["densidad pob"]=paises3.apply(densidad, axis=1)
paises3
```
